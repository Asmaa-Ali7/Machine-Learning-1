# `05` Logistic Regression

Using `sklearn` implementation of logistic regression to model the data.

Hint: Look for `sklearn.linear_model.LogisticRegression` class.

## `i` Baseline Model Training

Initializing the model with initial parameters, fitting it to the training data, and get a baseline model.

Use:
- Non-regularized model.
- `random_state` $= 1234$

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing

scaler = preprocessing.StandardScaler().fit(x_train)
x_scaled = scaler.transform(x_train)
```

```python
class_names = ['Poor', 'Standard', 'Good']
y_train.columns = y_train.columns.astype(str)
y_train = np.ravel(y_train)

model = LogisticRegression(random_state=1234, max_iter=30000).fit(x_scaled,y_train)
```

## `ii` Model Evaluation

Evaluate the baseline model performance using classification metrics, e.g. **Accuracy**, **Precision**, $F_1 \space Score$, **Confusion Matrix** etc.

Which metric is more reliable?

```python
print(classification_report(y_train, model.predict(x_scaled), target_names=class_names, digits=4,zero_division=0))
```

```python
#f1_score , No 
```

## `iii` Regularization Mode: On

Let's try the default $l_2$ regularization. Use default $C$ parameter for simplicity. Does regularization make the model generalize better?

Which of the two regularizations performs better?

```python
model_regularized= LogisticRegression(penalty='l2',C=1.0,random_state=1234,max_iter=30000).fit(x_scaled,y_train)
```

```python
print(classification_report(y_train, model_regularized.predict(x_scaled), target_names=class_names, digits=4,zero_division=0))
```

```python
model_regularized2= LogisticRegression(penalty='l2',random_state=1234, max_iter=30000).fit(x_scaled,y_train)
```

```python
print(classification_report(y_train, model_regularized2.predict(x_scaled), target_names=class_names, digits=4, zero_division=0))
```

```python
model_regularized3= LogisticRegression(C=1.0,random_state=1234, max_iter=30000).fit(x_scaled,y_train)
```

```python
print(classification_report(y_train, model_regularized2.predict(x_scaled), target_names=class_names, digits=4, zero_division=0))
```

## `iii` Tuning Hyperparameters

Try different `C` values and evaluate. What is the panalty value?

Remember that $alpha=\frac{1}{C}$ which means less regularization for higher $C$ values.

`Note:` You can tune other parameters as you see fit (but make sure you elaborate your reasoning.

Hint: You might want to use `sklearn.model_selection.RandomizedSearchCV` or `sklearn.model_selection.GridSearchCV` classes.

```python
from sklearn.model_selection import GridSearchCV


Model = LogisticRegression()

param_grid = { 'penalty':['l1', 'l2', 'elasticnet', None],'C':[.00005,.0005,0.005,0.002,.001,0.01], 'solver':['lbfgs', 'newton-cg','newton-cholesky','sag','saga']}

# Create the grid search with cross-validation

grid_search = GridSearchCV(estimator=Model, param_grid=param_grid, cv=10)
grid_search.fit(x_scaled, y_train)
best_params = grid_search.best_params_

print("Best Parameters:", best_params)
```

```python
grid_search.best_estimator_
```

# `06` Generalization Error

Now that we have tuned our solution, to an *extent*, we need to test for a final time on the held out test set.

```python
#x_test.columns = x_test.columns.astype(str)
x_test_scaled = scaler.transform(x_test)
y_test.columns = y_test.columns.astype(str)
print(classification_report(y_test, grid_search.best_estimator_.predict(x_test_scaled), target_names=class_names, digits=4, zero_division=0))
```
